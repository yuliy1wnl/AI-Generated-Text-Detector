```python
from flask import Flask, request, render_template, jsonify
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from pyngrok import ngrok

app = Flask(
    __name__,
    template_folder="/content/drive/MyDrive/assets/gui/templates",
    static_folder="/content/drive/MyDrive/assets/gui/static"
)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load model/tokenizer
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/assets/models/saved_modelv2")
model = AutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/assets/models/saved_modelv2")
model.to(device)
model.eval()

def predict(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
    prediction = torch.argmax(logits, dim=-1).item()
    return "ğŸ§  AI-generated" if prediction == 1 else "âœï¸ Human-written"

@app.route("/", methods=["GET"])
def index():
    # Serve the HTML page; no POST here anymore
    return render_template("index.html")

@app.route("/predict", methods=["POST"])
def predict_api():
    data = request.get_json()
    user_text = data.get("user_text", "")
    if not user_text.strip():
        return jsonify({"error": "No text provided"}), 400
    prediction = predict(user_text)
    return jsonify({"result": prediction})

# Ngrok setup
ngrok.set_auth_token("your_token")

public_url = ngrok.connect(5000)
print(f"ğŸ”— Your app is live at: {public_url}")

app.run(port=5000)
```
